```python
def rename_columns(df):
    rename_col = {
        'From Last': 'FromLast',
        'Same As Day': 'SameAsDay',
        'Winning Number 1': 'Num1',
        '2': 'Num2',
        '3': 'Num3',
        '4': 'Num4',
        '5': 'Num5',
        '6': 'Num6',
        'Additional Number': 'Num7',
        'Division 1 Winners': 'Div1Win',
        'Division 2 Winners': 'Div2Win',
        'Division 3 Winners': 'Div3Win',
        'Division 4 Winners': 'Div4Win',
        'Division 5 Winners': 'Div5Win',
        'Division 6 Winners': 'Div6Win',
        'Division 7 Winners': 'Div7Win'
    }
    return df.rename(columns=rename_col)

def my_one_hot_encoding(lst_of_lst):
    data = []
    for lst in lst_of_lst:
        one_hot = [0] * 49
        for el in lst:
            one_hot[el - 1] = 1
        data.append(one_hot)
    # print(data)
    return data

def my_inverse_one_hot_encoding(lst):
    return [i + 1 for i, val in enumerate(lst) if val == 1]
```

```python
import torch
import numpy as np
from torch.utils.data import DataLoader, TensorDataset

# Dummy data
data = np.random.rand(2000, 49)  # Increase dataset size

seq_length = 32
batch_size = 64
max_epochs = 100

X, y = [], []

for i in range(len(data) - seq_length):
    X.append(data[i:i+seq_length])
    y.append(data[i+seq_length])

X = np.array(X)  # Shape: (num_samples, seq_length, input_dim)
y = np.array(y)

# Convert to PyTorch tensors
X_tensor = torch.FloatTensor(X)
y_tensor = torch.LongTensor(np.argmax(y, axis=1))  # Keep categorical targets

# Create DataLoader
dataset = TensorDataset(X_tensor, y_tensor)
train_loader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

class PositionalEncoding(torch.nn.Module):
    def __init__(self, d_model, max_len=5000):
        super().__init__()
        pe = torch.zeros(max_len, d_model)
        position = torch.arange(max_len, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, d_model, 2).float() * (-np.log(10000.0) / d_model))
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        self.register_buffer('pe', pe)

    def forward(self, x):
        return x + self.pe[:x.size(1)].unsqueeze(0)  # Fix dimension alignment

class TransformerPredictor(torch.nn.Module):
    def __init__(self, input_dim=49, emb_dim=128, nhead=8, num_layers=4):
        super().__init__()
        self.emb_dim = emb_dim
        self.embedding = torch.nn.Linear(input_dim, emb_dim)
        self.pos_encoder = PositionalEncoding(emb_dim)
        encoder_layer = torch.nn.TransformerEncoderLayer(d_model=emb_dim, nhead=nhead)
        self.transformer = torch.nn.TransformerEncoder(encoder_layer, num_layers)
        self.alpha_predictor = torch.nn.Sequential(
            torch.nn.Linear(emb_dim, 49),
            torch.nn.Softplus()
        )

    def forward(self, src):
        src = self.embedding(src) * np.sqrt(self.emb_dim)  # Shape: (batch, seq_len, emb_dim)
        src = self.pos_encoder(src)
        src = src.permute(1, 0, 2)  # Transformer expects (seq_len, batch, emb_dim)
        output = self.transformer(src)
        return self.alpha_predictor(output[-1])  # Output last timestep for each batch

def dirichlet_loss(alphas, targets):
    alpha_0 = alphas.sum(dim=1)
    log_probs = torch.digamma(alphas[torch.arange(alphas.size(0)), targets]) - torch.digamma(alpha_0)
    return -log_probs.mean()

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = TransformerPredictor().to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

# Training loop
for epoch in range(max_epochs):
    model.train()
    total_loss = 0
    for batch_X, batch_y in train_loader:
        batch_X = batch_X.to(device)  # (batch, seq_len, input_dim)
        batch_y = batch_y.to(device)

        optimizer.zero_grad()
        alphas = model(batch_X)
        loss = dirichlet_loss(alphas, batch_y)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch+1}, Loss: {total_loss/len(train_loader):.4f}")
```

```python
def predict_next(model, sequence):
    model.eval()
    with torch.no_grad():
        seq_tensor = torch.FloatTensor(sequence).unsqueeze(1).permute(1, 0, 2)
        alphas = model(seq_tensor.to(device)).cpu().numpy()

    # Dirichlet distribution parameters
    alpha_0 = alphas.sum()
    mean_probs = alphas / alpha_0
    uncertainty = 1 / (alpha_0 + 1)  # Simple uncertainty measure

    return {
        'mean_probability': mean_probs,
        'uncertainty': uncertainty,
        'alphas': alphas
    }

# Example usage
test_sequence = X[467]  # Replace with actual sequence
prediction = predict_next(model, test_sequence)
print("Predicted probabilities:", prediction['mean_probability'])
print("Uncertainty:", prediction['uncertainty'])

prediction['alphas']
```
